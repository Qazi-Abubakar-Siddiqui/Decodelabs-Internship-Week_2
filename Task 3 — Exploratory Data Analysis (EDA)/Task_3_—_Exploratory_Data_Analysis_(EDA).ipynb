```python
# ============================================
# TASK 3 — EXPLORATORY DATA ANALYSIS (EDA)
# ============================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"

df = pd.read_csv(url)

numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)

print("Statistical Summary:")
print(df.describe())

print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))

print("\nTop Manufacturers:")
print(df['Manufacturer'].value_counts().head())

print("\nAverage Price by Manufacturer:")
print(df.groupby('Manufacturer')['Price'].mean().sort_values(ascending=False).head())

print("\nHighest Price:")
print(df['Price'].max())

print("\nLowest Price:")
print(df['Price'].min())

print("\nAverage Mileage:")
print(df['MPG.city'].mean())

sns.boxplot(x=df['Price'])
plt.title("Price Outliers")
plt.show()
```
